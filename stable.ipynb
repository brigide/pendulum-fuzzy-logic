```python
#imports
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
#Stable


#Variaveis de Entrada (Antecedent)
input1 = ctrl.Antecedent(np.arange(-30, 30, 0.5), 'input1')
input2 = ctrl.Antecedent(np.arange(-6, 6, 0.1), 'input2')
input3 = ctrl.Antecedent(np.arange(-0.4, 0.4, 0.05), 'input3')
input4 = ctrl.Antecedent(np.arange(-1, 1, 0.1), 'input4')

#Variaveis de saída (Consequent)
vector = ctrl.Consequent(np.arange(-6, 6, 0.2), 'vector')


#Atribuição de categorias
input1['nvb'] = fuzz.trapmf(input1.universe, [-30, -30, -18, -12])
input1['nb'] = fuzz.trimf(input1.universe, [-16.5, -10.5, -4.5])
input1['n'] = fuzz.trimf(input1.universe, [-9, -4.5, 0])
input1['zo'] = fuzz.trimf(input1.universe, [-3, 0, 3])
input1['p'] = fuzz.trimf(input1.universe, [0, 4.5, 9])
input1['pb'] = fuzz.trimf(input1.universe, [4.5, 10.5, 16.5])
input1['pvb'] = fuzz.trapmf(input1.universe, [12, 18, 30, 30])

input2['nb'] = fuzz.trapmf(input2.universe, [-6, -6, -4.2, -1.7])
input2['n'] = fuzz.trimf(input2.universe, [-3.6, -1.7, 0])
input2['zo'] = fuzz.trimf(input2.universe, [-1.7, 0, 1.7])
input2['p'] = fuzz.trimf(input2.universe, [0, 1.7, 3.6])   
input2['pb'] = fuzz.trapmf(input2.universe, [1.7, 4.2, 6, 6])

input3['nbig'] = fuzz.trapmf(input3.universe, [-0.4, -0.4, -0.3, -0.15])
input3['neg'] = fuzz.trimf(input3.universe, [-0.3, -0.15, 0])
input3['z'] = fuzz.trimf(input3.universe, [-0.15, 0, 0.15])
input3['pos'] = fuzz.trimf(input3.universe, [0, 0.15, 0.3])   
input3['pbig'] = fuzz.trapmf(input3.universe, [0.15, 0.3, 0.4, 0.4])

input4['neg'] = fuzz.trapmf(input4.universe, [-1, -1, -0.1, 0])
input4['z'] = fuzz.trimf(input4.universe, [-0.1, 0, 0.1]) 
input4['pos'] = fuzz.trapmf(input4.universe, [0, 0.1, 1, 1])


vector['nvvb'] = fuzz.trapmf(vector.universe, [-6, -6, -4.8, -3.6])
vector['nvb'] = fuzz.trimf(vector.universe, [-4.8, -3.6, -2.4])
vector['nb'] = fuzz.trimf(vector.universe, [-3.6, -2.4, -1.2])
vector['n'] = fuzz.trimf(vector.universe, [-2.4, -1.2, 0])
vector['z'] = fuzz.trimf(vector.universe, [-1.2, 0, 1.2])
vector['p'] = fuzz.trimf(vector.universe, [0, 1.2, 2.4])
vector['pb'] = fuzz.trimf(vector.universe, [1.2, 2.4, 3.6])
vector['pvb'] = fuzz.trimf(vector.universe, [2.4, 3.6, 4.8])
vector['pvvb'] = fuzz.trapmf(vector.universe, [3.6, 4.8, 6, 6])
```

```python
#visualização das variaveis
input1.view()
input2.view()
input3.view()
input4.view()
vector.view()
```

```python
#regras

rule1 = ctrl.Rule(input3['nbig'] & input4['neg'], vector['pvvb'])
rule2 = ctrl.Rule(input3['neg'] & input4['neg'], vector['pvb'])
rule3 = ctrl.Rule(input3['z'] & input4['neg'], vector['pb'])
rule4 = ctrl.Rule(input3['z'] & input4['z'], vector['z'])
rule5 = ctrl.Rule(input3['z'] & input4['pos'], vector['nb'])
rule6 = ctrl.Rule(input3['pos'] & input4['pos'], vector['nvb'])
rule7 = ctrl.Rule(input3['pbig'] & input4['pos'], vector['nvvb'])

rule8 = ctrl.Rule(input1['nvb'] & input2['nb'], vector['nvvb'])
rule9 = ctrl.Rule(input1['nvb'] & input2['n'], vector['nvvb'])
rule10 = ctrl.Rule(input1['nvb'] & input2['zo'], vector['nvb'])
rule11 = ctrl.Rule(input1['nvb'] & input2['p'], vector['nb'])
rule12 = ctrl.Rule(input1['nvb'] & input2['pb'], vector['n'])

rule13 = ctrl.Rule(input1['nb'] & input2['nb'], vector['nvvb'])
rule14 = ctrl.Rule(input1['nb'] & input2['n'], vector['nvb'])
rule15 = ctrl.Rule(input1['nb'] & input2['zo'], vector['nb'])
rule16 = ctrl.Rule(input1['nb'] & input2['p'], vector['n'])
rule17 = ctrl.Rule(input1['nb'] & input2['pb'], vector['z'])

rule18 = ctrl.Rule(input1['n'] & input2['nb'], vector['nvb'])
rule19 = ctrl.Rule(input1['n'] & input2['n'], vector['nb'])
rule20 = ctrl.Rule(input1['n'] & input2['zo'], vector['n'])
rule21 = ctrl.Rule(input1['n'] & input2['p'], vector['z'])
rule22 = ctrl.Rule(input1['n'] & input2['pb'], vector['p'])

rule23 = ctrl.Rule(input1['zo'] & input2['nb'], vector['nb'])
rule24 = ctrl.Rule(input1['zo'] & input2['n'], vector['n'])
rule25 = ctrl.Rule(input1['zo'] & input2['zo'], vector['z'])
rule26 = ctrl.Rule(input1['zo'] & input2['p'], vector['p'])
rule27 = ctrl.Rule(input1['zo'] & input2['pb'], vector['pb'])

rule28 = ctrl.Rule(input1['p'] & input2['nb'], vector['n'])
rule29 = ctrl.Rule(input1['p'] & input2['n'], vector['z'])
rule30 = ctrl.Rule(input1['p'] & input2['zo'], vector['p'])
rule31 = ctrl.Rule(input1['p'] & input2['p'], vector['pb'])
rule32 = ctrl.Rule(input1['p'] & input2['pb'], vector['pvb'])

rule33 = ctrl.Rule(input1['pb'] & input2['nb'], vector['z'])
rule34 = ctrl.Rule(input1['pb'] & input2['n'], vector['p'])
rule35 = ctrl.Rule(input1['pb'] & input2['zo'], vector['pb'])
rule36 = ctrl.Rule(input1['pb'] & input2['p'], vector['pvb'])
rule37 = ctrl.Rule(input1['pb'] & input2['pb'], vector['pvvb'])

rule38 = ctrl.Rule(input1['pvb'] & input2['nb'], vector['p'])
rule39 = ctrl.Rule(input1['pvb'] & input2['n'], vector['pb'])
rule40 = ctrl.Rule(input1['pvb'] & input2['zo'], vector['pvb'])
rule41 = ctrl.Rule(input1['pvb'] & input2['p'], vector['pvvb'])
rule42 = ctrl.Rule(input1['pvb'] & input2['pb'], vector['pvvb'])

controller = ctrl.ControlSystem([
        rule1, rule2, rule3, rule4, rule5, rule6, rule7,
        rule8, rule9, rule10, rule11, rule12, rule13, rule14,
        rule15, rule16, rule17, rule18, rule19, rule20, rule21,
        rule22, rule23, rule24, rule25, rule26, rule27, rule28,
        rule29, rule30, rule31, rule32, rule33, rule34, rule35,
        rule36, rule37, rule38, rule39, rule40, rule41, rule42
    ])
```

```python
#Simulação
calculateVector = ctrl.ControlSystemSimulation(controller)

inputInput1 = 10 #int(input("angle (30, 330): "))
inputInput2 = 5 #int(input("speed (-10, 10): "))
inputInput3 = 0.05 #int(input("angle (30, 330): "))
inputInput4 = 0 #int(input("speed (-10, 10): "))

calculateVector.input['input1'] = inputInput1
calculateVector.input['input2'] = inputInput2
calculateVector.input['input3'] = inputInput3
calculateVector.input['input4'] = inputInput4
calculateVector.compute()
```

```python
#output
vectorOutput = calculateVector.output['vector']

print("\nInput1 %d \nInput2 %.1f \nInput3 %.1f \nInput4 %.1f \nVector %.1f" %(
        inputInput1,
        inputInput2,
        inputInput3,
        inputInput4,
        vectorOutput))

input1.view(sim=calculateVector)
input2.view(sim=calculateVector)
input3.view(sim=calculateVector)
input4.view(sim=calculateVector)
vector.view(sim=calculateVector)

plt.show()
```
